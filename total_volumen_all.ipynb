```python
from binance.client import Client
import pandas as pd
import datetime

# Reemplaza con tus propias claves API
api_key = '3WiCAiEyWi7HjnBd6RVgpq9P7AK9STyrW7aaC5cjwkfWxX9LkrHruGRbbrCKmROv'
api_secret = 'am3Lb7vnJHcYZosuf3CPLgSyXKkJKIEjrN4rAJ2469IMbm5t9dJxBQCnyj4VIVhl'

# Crear una instancia del cliente Binance
client = Client(api_key, api_secret)

# Obtener la lista de todos los pares de trading disponibles
exchange_info = client.get_exchange_info()
symbols = [symbol['symbol'] for symbol in exchange_info['symbols']]

# Obtener la fecha actual y la fecha hace 5 años
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=5 * 365)  # 5 años atrás

# Crear un DataFrame vacío para almacenar los datos
data = []

# Iterar sobre los días
for day in range((end_date - start_date).days + 1):
    current_date = start_date + datetime.timedelta(days=day)
    
    try:
        total_daily_volume = 0
        
        # Obtener los datos de klines (candlesticks) para cada día en cada par
        for symbol in symbols:
            klines = client.get_historical_klines(symbol, Client.KLINE_INTERVAL_1DAY, current_date.strftime('%d %b %Y %H:%M:%S'))

            # Calcular el volumen total diario sumando los volúmenes de cada día
            total_daily_volume += sum(float(kline[5]) for kline in klines)
            
        data.append([current_date.date(), total_daily_volume])
        
    except Exception as e:
        print(f"Error al obtener datos para {current_date}: {str(e)}")

# Crear el DataFrame final
df = pd.DataFrame(data, columns=['Fecha', 'Volumen total diario'])

# Guardar el DataFrame en un archivo CSV
df.to_csv('E:/sabas/Henry/Proyecto2/PI_DA/Dataset/volumen_total_diario_binance.csv', index=False)

print("Datos exportados a volumen_total_diario_binance.csv")
```
