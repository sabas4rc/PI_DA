```python
import mysql.connector
import pandas as pd
from datetime import datetime

# Conexión a la base de datos MySQL
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678",
    database="cripto_db"
)
```

```python
df = pd.read_csv('historial_rendimiento_bonos_tesoro.csv')
```

```python
# Cambiar los nombres de las columnas
df.rename(columns={'Date': 'fecha', 'Open': 'open','High': 'high','Low': 'low','Close': 'close'}, inplace=True)
```

```python
df.head(10)
```

```python
# Crear una instancia del cursor
cursor = db_connection.cursor()

# Insertar los datos desde el DataFrame a la tabla 'spread_cripto'
insert_query = "INSERT INTO bonos_usa (fecha, open, high, low, close) VALUES (%s,%s,%s,%s,%s)"
for index, row in df.iterrows():
    values = (row['fecha'],row['open'],row['high'],row['low'],row['close'])
    cursor.execute(insert_query, values)

db_connection.commit()

# Cerrar la conexión
cursor.close()
db_connection.close()

print("Datos importados a la tabla 'spread_cripto' en la base de datos MySQL")
```
